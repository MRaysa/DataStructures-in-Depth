#Problem 1
Take a string from the user and print it in the reverse order using stack
data structure. Also prints the unicode of each character

Input:
W A T E R M E L O N

Output:
N O L E M R E T A W

Hint: Use the property of stack to reverse the string

```python
class Stack:
    top = 0
    stack_size = 0
    stack = []

    def __init__(self, size):
        self.stack_size = size
        self.stack = [0] * size

    def stack_is_full(self):
        if self.top == self.stack_size:
            return True
        else:
            return False

    def push(self, value):
        if self.stack_is_full():
            print("Stack is full")
        else:
            self.stack[self.top] = value
            self.top = self.top + 1

    def stack_is_empty(self):
        if self.top == 0:
            return True
        else:
            return False

    def peak(self):
        if self.stack_is_empty():
            print("Stack is empty")
            return 0
        else:
            return self.stack[self.top - 1]

    def pop(self):
        if self.stack_is_empty():
            print("Stack is empty")
        else:
            self.stack[self.top - 1] = 0
            self.top = self.top - 1

def reverse(string):
    n = len(string)
    s = Stack(n)
    output = ""

    for i in range(n):
        s.push(string[i])

    while not s.stack_is_empty():
        output = output + s.peak()

        s.pop()

    return output

string = input("Enter a string : ")
string = reverse(string)
print("Reversed string is : "  + string)
```

#Problem 2.  
 Given a string of brackets, determine whether each sequence of brackets is
balanced. If a string is balanced, print True on a new line; otherwise, print False on
a new line.
Input : Output

()  True

{} True

[] True

{[]} True

{([)]} False

(() False

{[(]} False

```python
class Stack:
    top = 0
    stack_size = 0
    stack = []

    def __init__(self, size):
        self.stack_size = size
        self.stack = [0] * size

    def stack_is_full(self):
        if self.top == self.stack_size:
            return True
        else:
            return False

    def push(self, value):
        if self.stack_is_full():
            print("Stack is full")
        else:
            self.stack[self.top] = value
            self.top = self.top + 1

    def stack_is_empty(self):
        if self.top == 0:
            return True
        else:
            return False

    def peak(self):
        if self.stack_is_empty():
            print("Stack is empty")
            return 0
        else:
            return self.stack[self.top - 1]


    def pop(self):
        if self.stack_is_empty():
            print("Stack is empty")
        else:
            self.stack[self.top - 1] = 0
            self.top = self.top - 1

    def find_Balance(each_bracket):
        stack = []
        for char in each_bracket:
            if char in ["(", "{", "["]:
                stack.append(char)
            else:
                if not stack:
                    return False
                curr_char = stack.pop()
                if curr_char == '(':
                    if char != ")":
                        return False
                if curr_char == '{':
                    if char != "}":
                        return False
                if curr_char == '[':
                    if char != "]":
                        return False

        if stack:
            return False
        return True


    m = input("Bracket: ")

    if find_Balance(m):
        print("True")
    else:
        print("False")
```

#Problem 4.

the stack by adding a new element in any of the stack. For example, if stack A contains 1,
5, 11 and 4 and stack B contains 4 ,32,- 1 and 1 then by adding 15 in stack A will balance
the sum of values of both the stacks.
Input:
1 4

11

Stack A Stack B
Output:
1 4

11 1

Stack A 15

```python
class Stack:
    top = 0
    stack_size = 0
    stack = []

    def __init__(self, size):
        self.stack_size = size
        self.stack = [0] * size

    def is_full(self):
        if self.top == self.stack_size:
            return True
        else:
            return False

    def push(self, value):
        if self.is_full():
            print("Stack is full")
        else:
            self.stack[self.top] = value
            self.top = self.top + 1

    def is_empty(self):
        if self.top == 0:
            return True
        else:
            return False

    def peak(self):
        if self.is_empty():
            print("Stack is empty")
        else:
            return self.stack[self.top - 1]

    def pop(self):
        if self.is_empty():
            print("Stack is empty")
        else:
            self.stack[self.top - 1] = 0
            self.top = self.top - 1


elements = int(input("How many elements in the stack?: "))
stackA = Stack(elements + 1)
stackB = Stack(elements + 1)

total_A = 0
total_B = 0

for i in range(elements):
    element_A = int(input("Enter element for stack A: "))
    element_B = int(input("Enter element for stack B: "))

    total_A = total_A + element_A
    total_B = total_B + element_B

    stackA.push(element_A)
    stackB.push(element_B)


print("Before balancing: ")
print("Stack A:", stackA.stack)
print("Stack B:", stackB.stack)

sum_of_ellements = abs(total_B - total_A)

if total_A < total_B:
    stackA.push(sum_of_ellements)
else:
    stackB.push(sum_of_ellements)


print("After balancing: ")
print("Stack A:", stackA.stack)
print("Stack B:", stackB.stack)
```

```python
class Stack:
    top = 0
    stack_size = 0
    stack = []

    def __init__(self, size):
        self.stack_size = size
        self.stack = [0] * size

    def is_full(self):
        if self.top == self.stack_size:
            return True
        else:
            return False

    def push(self, value):
        if self.is_full():
            print("Stack is full")
        else:
            self.stack[self.top] = value
            self.top = self.top + 1

    def is_empty(self):
        if self.top == 0:
            return True
        else:
            return False

    def peak(self):
        if self.is_empty():
            print("Stack is empty")
            return 0
        else:
            return self.stack[self.top - 1]

    def pop(self):
        if self.is_empty():
            print("Stack is empty")
        else:
            self.stack[self.top - 1] = 0
            self.top = self.top - 1

    def pushStack(self, value):
        if self.is_full():
            print("Stack is full")
        else:
            if self.is_empty():
                self.push(value)
            else:
                temp_stack = Stack(self.stack_size)

                while not ((self.is_empty()) or (value > self.peak())):
                    top = self.peak()
                    temp_stack.push(top)
                    self.pop()

                self.push(value)

                while not temp_stack.is_empty():
                    top = temp_stack.peak()
                    self.push(top)
                    temp_stack.pop()


stack1 = Stack(5)

stack1.pushStack(160)
stack1.pushStack(67)
stack1.pushStack(456)
stack1.pushStack(259)
stack1.pushStack(120)

print(stack1.stack)
```

#problem 7.
Reverse the first âkâ elements of a queue, push it back into the queue. For example, if the
queue has members 10, 20, 30, 40 , 50 , 60 , 70, 80, 90 and first 5 numbers are asked to be
reversed then the result will be 60, 70, 80, 90, 50, 40, 30, 20, 10.

Input:

10 20 30 40 50 60 70 80 90

Output:

60 70 80 90 50 40 30 20 10

```python
class Queue:
    def __init__(self):
        self.queue = []

    def size(self):
        return len(self.queue)

    def is_empty(self):
        if self.size() == 0:
            return True
        else:
            return False

    def enqueue(self, data):
        self.queue.append(data)


    def dequeue(self):
        if self.is_empty():
            print("Queue is empty")
        else:
            self.queue.pop(0)

    def front(self):
        if self.is_empty():
            print("Queue is empty")
        else:
            return self.queue[0]

    def back(self):
        if self.is_empty():
            print("Queue is empty")
        else:
            return self.queue[-1]


class Stack:
    def __init__(self):
        self.stack = []

    def size(self):
        return len(self.stack)

    def is_empty(self):
        if self.size() == 0:
            return True
        else:
            return False

    def push(self, data):
        self.stack.append(data)


    def pop(self):
        if self.is_empty():
            print("Stack is empty")
        else:
            self.stack.pop(-1)

    def peak(self):
        if self.is_empty():
            print("Stack is empty")
        else:
            return self.stack[-1]


num_elements = int(input("How many elements in the queue: "))

q = Queue()
s = Stack()
for i in range(num_elements):
    element = int(input("Enter element: "))
    q.enqueue(element)

print("Before reverse:", q.queue)

while True:
    num = int(input("How many numbers to reverse?: "))

    if num <= q.size():
        break
    else:
        print("Cannot reverse more than the size of the queue")


for i in range(num):
    s.push(q.front())
    q.dequeue()

while not s.is_empty():
    q.enqueue(s.peak())
    s.pop()

print("After Reverse:", q.queue)
```

whose ids are divisible by 2 and volunteer B is collecting donations from members whose
ids are not divisible by 2. Write a program which uses two queues, such that if the even
